### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()
```

```python
print(ingreso_total_tienda1)
print(ingreso_total_tienda2)
print(ingreso_total_tienda3)
print(ingreso_total_tienda4)
```

# 2. Ventas por categoría

```python
ventas_por_categoria1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_por_categoria2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_por_categoria3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_por_categoria4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
```

```python
print(ventas_por_categoria1)
print(ventas_por_categoria2)
print(ventas_por_categoria3)
print(ventas_por_categoria4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio1 = tienda['Calificación'].mean()
calificacion_promedio2 = tienda2['Calificación'].mean()
calificacion_promedio3 = tienda3['Calificación'].mean()
calificacion_promedio4 = tienda4['Calificación'].mean()
```

```python
print(calificacion_promedio1)
print(calificacion_promedio2)
print(calificacion_promedio3)
print(calificacion_promedio4)
```

# 4. Productos más y menos vendidos

```python
producto_mas_vendido1 = tienda.groupby('Producto')['Producto'].count().sort_values(ascending=False)
producto_mas_vendido2 = tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False)
producto_mas_vendido3 = tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False)
producto_mas_vendido4 = tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False)
```

# 6. Graficos


```python
import numpy as np
import matplotlib.ticker as mticker

fig, ax = plt.subplots(figsize=(12, 6)) # Increased figure size for better visibility

# Example data
tiendas = ('Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4')
y_pos = np.arange(len(tiendas))
ganancias = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]

bars = ax.barh(y_pos, ganancias, align='center')
ax.set_yticks(y_pos, labels=tiendas)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Ingresos')
ax.set_title('Ingresos por tienda')

# Format x-axis labels to show values in billions
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(x/1000000000, '.1f') + 'B'))

# Set the major tick locations to every 0.05 Billion (50 million) for clearer intervals
ax.xaxis.set_major_locator(mticker.MultipleLocator(100000000)) # 50,000,000 is 0.05 Billion

# Set the x-axis limits to emphasize differences
ax.set_xlim(left=1000000000) # Setting the lower limit to 1 billion (1,000,000,000)

# Add the values of each bar inside the bar
for idx, bar in enumerate(bars):
    width = bar.get_width()
    ax.text(width - 10000000, bar.get_y() + bar.get_height()/2,    # (x,y) position
            f'{str(ganancias[idx]/1000000000)[:4] + 'B'}',                                   # format to 2 decimal place
            ha='right', va='center', color='white', weight='bold') # text alignment and style

plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

fig, (ax, ax2, ax3, ax4) = plt.subplots(nrows=1,ncols=4,figsize=(16, 12), subplot_kw=dict(aspect="equal"))
#fig, ax2 = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

categorias = ventas_por_categoria1.index
valores = ventas_por_categoria1.values
valores2 = ventas_por_categoria2.values
valores3 = ventas_por_categoria3.values
valores4 = ventas_por_categoria4.values
explode = (0.05, 0, 0, 0, 0,0,0,0)
cmap_colors = cm.tab10.colors[:len(categorias)]

def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return f"{pct:.1f}%\n({absolute})"

wedges, texts, autotexts = ax.pie(valores,colors=cmap_colors, explode=explode, autopct=lambda pct: func(pct, valores), textprops=dict(color="w"))
wedges, texts, autotexts2 = ax2.pie(valores2,colors=cmap_colors, explode=explode, autopct=lambda pct: func(pct, valores2), textprops=dict(color="w"))
wedges, texts, autotexts3 = ax3.pie(valores3,colors=cmap_colors, explode=explode, autopct=lambda pct: func(pct, valores3), textprops=dict(color="w"))
wedges, texts, autotexts4 = ax4.pie(valores4,colors=cmap_colors, explode=explode, autopct=lambda pct: func(pct, valores4), textprops=dict(color="w"))

ax.legend(wedges, categorias,
          title="Categorias",
          loc="center left",
          bbox_to_anchor=(5, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
plt.setp(autotexts2, size=8, weight="bold")
plt.setp(autotexts3, size=8, weight="bold")
plt.setp(autotexts4, size=8, weight="bold")

ax.set_title("Ventas por categoria Tienda 1")
ax2.set_title("Ventas por categoria Tienda 2")
ax3.set_title("Ventas por categoria Tienda 3")
ax4.set_title("Ventas por categoria Tienda 4")

plt.show()
```

```python
import numpy as np
import matplotlib.ticker as mticker

fig, ax = plt.subplots(figsize=(12, 6))

tiendas = ('Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4')
puntuaciones = [calificacion_promedio1, calificacion_promedio2,
                calificacion_promedio3, calificacion_promedio4]
bar_colors = ['tab:blue', 'tab:orange', 'tab:red', 'tab:blue']

bars = ax.bar(tiendas, puntuaciones, color=bar_colors)

ax.set_ylabel('Promedio Puntuacion')
ax.set_title('Promedio Puntuacion por Tienda')
ax.set_ylim(bottom=3, top=5)

# Add the values of each bar inside the bar
for idx, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height - 0.1,    # (x,y) position
            f'{str(puntuaciones[idx])[:4]}',
            ha='center', va='center', color='white', weight='bold')  # text alignment and style


plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.ticker as mticker

fig, ax = plt.subplots(figsize=(6, 6)) # Increased figure size for better visibility

# Example data
productos = pd.concat([producto_mas_vendido1.tail(5).sort_values(),producto_mas_vendido1.head(5).sort_values()])
nombres_productos = productos.index
y_pos = np.arange(len(nombres_productos))
puntuaciones = productos.values

bars = ax.barh(y_pos, puntuaciones, align='center')
ax.set_yticks(y_pos, labels=nombres_productos)
ax.set_xlabel('Ventas')
ax.set_title('Productos mas y menos vendidos Tienda 1')

# Add the values of each bar inside the bar
for idx, bar in enumerate(bars):
    width = bar.get_width()
    ax.text(width - 1, bar.get_y() + bar.get_height()/2,    # (x,y) position
           f'{str(puntuaciones[idx])[:4]}',                                   # format to 1 decimal place
           ha='right', va='center', color='white', weight='bold') # text alignment and style

plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```
